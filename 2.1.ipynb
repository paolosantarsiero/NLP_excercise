# Extractive Text Summarization 

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
import urllib
import bs4 as BeautifulSoup
import re
```

```python
scraped_data = urllib.request.urlopen('https://en.wikipedia.org/wiki/Artificial_intelligence')
article = scraped_data.read()

parsed_article = BeautifulSoup.BeautifulSoup(article,'html.parser')

paragraphs = parsed_article.find_all('p')

article_text = ""

for p in paragraphs:
    article_text += p.text

article_text = re.sub(r'\[[0-9]*\]', ' ', article_text)
article_text = re.sub(r'\s+', ' ', article_text)
formatted_article_text = re.sub('[^a-zA-Z]', ' ', article_text )
formatted_article_text = re.sub(r'\s+', ' ', formatted_article_text)
```

```python
print(formatted_article_text)
```

```python
stop_words = set(stopwords.words('english'))
words = word_tokenize(formatted_article_text)
freqTable = dict()

for word in words:
    word = word.lower()
    if word not in stop_words:
        if word not in freqTable.keys():
            freqTable[word] = 1
        else:
            freqTable[word] = freqTable[word] + 1

max_frequency = max(freqTable.values())
for word in freqTable.keys():
    freqTable[word] = freqTable[word]/max_frequency

sents = sent_tokenize(article_text)
print(sents)
scoredSents = dict()

for sent in sents:
    for word in sent:
        word = word.lower()        
        if word in freqTable.keys():
            if sent not in scoredSents.keys():
                scoredSents[sent] = freqTable[word]
            else:
                scoredSents[sent] = scoredSents[sent] + freqTable[word]
```

```python
select_length = int(len(sents)*0.3)
sorted_sents = dict(sorted(scoredSents.items(), key=lambda item: item[1], reverse=True));
summary = " "

keys_list = list(sorted_sents)
for i in range(0, select_length):
    sent = keys_list[i]
    print(sent)
    summary.join(sent)
```

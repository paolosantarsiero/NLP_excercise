# Word Sense Induction

```python
import nltk
nltk.download('brown')

sents = nltk.corpus.brown.sents()
words = nltk.corpus.brown.words()
```

```python
fd = nltk.FreqDist(words)
nltk.corpus.brown.words()
for word, freq in fd.items():
    if freq >= 50 and freq <= 1000:
            print(word, freq)
```
